```python
# import required libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# load the train dataset
df_1 = pd.read_csv('../data/train.csv')
df_1
```

```python
# load the test dataset
df_2 = pd.read_csv('../data/test.csv')
df_2
```

<h3 style="color:blue"><b> 1. CONCAT <b> </h3>

```python
# concatenate the two datasets vertically
df_full = pd.concat([df_1, df_2], axis=0)
# reset the index
df_full.reset_index(drop=True, inplace=True)
df_full
```

```python
### fill the missing values with random values of the same column
def fill_missing(col):
    non_nan_values = col.dropna()
    return col.apply(lambda x: x if pd.notnull(x) else np.random.choice(non_nan_values))

# fill the missing values in the columns
df_full['target_carbon_monoxide'] = fill_missing(df_full['target_carbon_monoxide'])
df_full['target_benzene'] = fill_missing(df_full['target_benzene'])
df_full['target_nitrogen_oxides'] = fill_missing(df_full['target_nitrogen_oxides'])
df_full = df_full.copy()
df_full
```

```python
# save the data
df_full.to_csv('../data/df_full.csv', index=False)
```

<h3 style="color:blue"><b> 2. FILTER </b></h3>

```python
# Ensure the date_time column is in datetime format
df_full['date_time'] = pd.to_datetime(df_full['date_time'])
```

```python
# Filter for weekdays (0 = Monday to 4 = Friday), relative humidity > 50, and absolute humidity < 1.4
df_filtered = df_full[
    (df_full['date_time'].dt.weekday <= 4) &  # Weekdays only
    (df_full['relative_humidity'] > 50) &  # Relative humidity > 50
    (df_full['absolute_humidity'] < 1.4)  # Absolute humidity < 1.4
].copy()

# Reset the index
df_filtered.reset_index(drop=True, inplace=True)
df_filtered
```

<h3 style="color:blue"><b> 3. APPLY </b></h3>

```python
# A new(dummy) column for year-month to help with grouping and averaging of monthly sensor_4 values
# I would remove this column before saving
df_filtered['year_month'] = df_filtered['date_time'].dt.to_period('M')  # Year-Month as 'YYYY-MM'
```

```python
# Mean of 'sensor_4' for each year-month
monthly_mean_sensor_4 = df_filtered.groupby('year_month')['sensor_4'].transform('mean')
```

```python
monthly_mean_sensor_4
```

```python
# median of 'sensor_4'
sensor_4_median = df_filtered['sensor_4'].median()
sensor_4_median
```

```python
# creating column with name 'apply_step3' based on the condition
df_filtered['apply_step3'] = np.where(
    df_filtered['sensor_4'] < sensor_4_median,
    df_filtered['sensor_2'] * df_filtered['sensor_3'],  # If sensor_4 < median
    df_filtered['sensor_2'] * df_filtered['sensor_4'] - monthly_mean_sensor_4  # Else
)
```

```python
# drop the 'year_month' (dummy) column
df_filtered.drop(columns=['year_month'], inplace=True)
df_filtered
```

```python
# save df_filtered table
df_filtered.to_csv('../data/df_filtered.csv', index=False)
```

<h3 style="color:blue"><b> 5. Create a column in df_full </b></h3>

```python
# A function to map values to classs using a dictionary
def map_to_classs_from_dict(value, split_dict):
    """
    Map a value to a class based on a dictionary of splits.
    Args:
        value: The input value to be mapped to a class.
        split_dict: A dictionary where keys are class labels and values are class ranges.
                  Example: {"class_1": [0, 0.2], "class_2": [0.2, 0.5]}
    Returns:
        The split class the value belongs to.
    """
    for class_label, class_range in split_dict.items():
        if class_range[0] <= value < class_range[1]:  # Check if value falls in the class range
            return class_label
    # Handle the edge case where value equals the upper boundary of the last class
    last_class = list(split_dict.values())[-1]
    if value == last_class[1]:
        return list(split_dict.keys())[-1]
    
# Define the class dictionary
split_dict = {
    "class_1": [0, 0.2],
    "class_2": [0.2, 0.5],
    "class_3": [0.5, 0.8],
    "class_4": [0.8, 0.85],
    "class_5": [0.85, 1],
}

# Normalize the columns
df_full["sensor_1_norm"] = df_full["sensor_1"] / df_full["sensor_1"].max()
df_full["sensor_2_norm"] = df_full["sensor_2"] / df_full["sensor_2"].max()

# Apply the function to create new columns
df_full["Truth"] = df_full["sensor_1_norm"].apply(map_to_classs_from_dict, split_dict=split_dict)
df_full["Predicted"] = df_full["sensor_2_norm"].apply(map_to_classs_from_dict, split_dict=split_dict)

df_full.head()
```

<h3 style="color:blue"><b> 6. Merge </b></h3>

```python
df_full.head()
```

```python
# Add 'month_day' column to represent the day and month (ignoring year)
df_full['month_day'] = df_full['date_time'].dt.strftime('%m-%d')  # Format: 'MM-DD'
df_full['month_day']
```

```python
# Filter for the year 2010 and 2011
df_year1 = df_full[df_full['date_time'].dt.year == 2010]
df_year2 = df_full[df_full['date_time'].dt.year == 2011]
```

```python
df_year2.head()
```

<span style="color:red">**Overlapp days between the two years (2010 and 2011)**</span>

```python
# Unique days in the year 2010
unique_days_year1 = df_year1['month_day'].unique()
unique_days_year1
```

```python
# Unique days in the year 2011
unique_days_year2 = df_year2['month_day'].unique()
unique_days_year2
```

```python
# overlapping days
overlapping_days = set(unique_days_year1).intersection(unique_days_year2)
overlapping_days
```

```python
# Number of overlapping days
print(f'Number of overlapping days: {len(overlapping_days)}')
```

<span style="color:red">**Daily difference average of absolute humidity**</span>

<span>The overlapping days have varying numbers of entries in each year, with multiple daily values recorded. To accurately compare daily values across years for these overlapping days, we first compute the daily average for each day in both years</span>

```python
# Aggregate absolute_humidity by day for each year
df_year1_daily = df_year1.groupby('month_day', as_index=False)['absolute_humidity'].mean()
df_year2_daily = df_year2.groupby('month_day', as_index=False)['absolute_humidity'].mean()
```

```python
df_year1_daily.head()
```

```python
df_year2_daily.head()
```

```python
# Merge the daily averages for overlapping days
df_daily_merged = pd.merge(
    df_year1_daily,
    df_year2_daily,
    on='month_day',
    suffixes=('_year1', '_year2')
)

print(len(df_daily_merged))
df_daily_merged.head()
```

```python
# daily difference in absolute humidity
df_daily_merged['abs_humidity_diff'] = abs(df_daily_merged['absolute_humidity_year1'] - df_daily_merged['absolute_humidity_year2'])

df_daily_merged.head()
```

```python
# average of the daily differences
avg_abs_humidity_diff = df_daily_merged['abs_humidity_diff'].mean()

print(f"Average daily difference of absolute humidity: {avg_abs_humidity_diff:.2f}")
```

```python
# df_1 is the original train dataset
df_1['date_time'] = pd.to_datetime(df_1['date_time'])
# extracting the day as unique days
df_1['day'] = df_1['date_time'].dt.date
unique_days = df_1['day'].unique()
np.random.seed(42)  # Set seed for reproducibility
# split into train and val having unique with train having 70%
train_days = np.random.choice(unique_days, size=int(0.7 * len(unique_days)), replace=False)  # 70% random sample
val_days = np.setdiff1d(unique_days, train_days)  # Remaining 30% for validation
```

```python
# Create the 'train_val_set' column based on the assigned days
df_1['train_val_set'] = df_1['day'].apply(lambda x: 'train' if x in train_days else 'val')
df_1.head()
```
